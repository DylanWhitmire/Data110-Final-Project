```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Data/Housing_Prices_24.csv")
df = df.drop('Id', axis=1)
```

```python
df
```

```python
print((df['KitchenQual']=='TA').sum())
```

```python
onlyNum = df.drop(columns=['LotConfig','BldgType','Foundation','KitchenQual'])
onlyNum = onlyNum.replace({'CentralAir':{'Y':1, 'N':0},'PavedDrive':{'Y':1, 'N':0}})
onlyNum
```

```python
corr = onlyNum.drop(columns=['SalePrice']).corr()


plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Housing Variables')
plt.show()
```
